```python
"""
Ildebrando Magnani

im975

02/16/2016

Homework#2

This program uses a bisection algorithm to solve for equilibrium (clearing) prices and quantities. It iterates the
procedure over four different sets of supply and demand curves' parameters (a, b, epsilon). The final output is 
displayed in a table that has results for Clearing Price, Equilibrium Demand, Equilibrium Supply and
Equilibrium Excess Supply. 

"""

import numpy as np
import pandas


def yd(price):
    global a, eps
    yd = a * (price**(-eps))
    return yd

def ys(price):
    global b
    ys = np.exp(b*price) - 1
    return ys


def rootfinder():
    alist = [1, 2, 1, 1]
    blist = [0.1, 0.1, 0.2, 0.1]
    epslist = [1, 1, 1, 2]
    global a, b, eps
    global clearprice_values
    global demand_values 
    global supply_values 
    global excesssupply_values
    clearprice_values = [] 
    demand_values = []
    supply_values = []
    excesssupply_values = []
    
    for x, y, z in zip(alist, blist, epslist):
        a = x
        b = y
        eps = z
        
        maxi = 40
        tol = 1.0e-12
        phigh = 10.0
        plow = 0.1
        ni = maxi
        
        for i in range(maxi):
            c = (plow + phigh)/2

            demand = yd(c)
            supply = ys(c)

            excesssupply = supply - demand

            if excesssupply > 0:
                phigh = c
            else:
                plow = c

            diff = np.abs(phigh - plow)

            if diff <= tol:
                ni = i
                break

            clearprice = (plow + phigh)/2    
            demand = yd(c)
            supply = yd(c)
            excesssupply = demand - supply  
        
        clearprice_values.append(clearprice)
        demand_values.append(demand)
        supply_values.append(supply)
        excesssupply_values.append(excesssupply)
    
    return clearprice_values

rootfinder()

x = np.array([[clearprice_values[0], clearprice_values[1], clearprice_values[2], clearprice_values[3]],
              [demand_values[0], demand_values[1], demand_values[2], demand_values[3]],
              [supply_values[0], supply_values[1], supply_values[2], supply_values[3]],
              [excesssupply_values[0], excesssupply_values[1], excesssupply_values[2], excesssupply_values[3]]])


col_labels = ['a=1, b=0.1 eps=1', 'a=2, b=0.1, eps=1', 'a=1, b=0.2, eps=1', 'a=1, b=0.1, eps=2']
row_labels = ['Clearing Price', 'Equilibrium Demand', 'Equilibrium Supply', 'Excess Supply in Eq']
table = pandas.DataFrame(x, columns = col_labels, index = row_labels)

table
```
