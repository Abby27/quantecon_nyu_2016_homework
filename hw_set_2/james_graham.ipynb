```python
# %load james_graham.py
#------------------------------------
# Homework Set 2
#
# James Graham, jag912, N11252710
#
# This file replicates in Python the supply and demand functions produced by 
# Tony Smith in Matlab. Uses bisection method to find the equilibrium quantity, y,
# and price, p. 
#
#------------------------------------

import math
import matplotlib.pyplot as plt
import numpy as np


def demand(price, a, epsilon):
    yd = a*(price**(-epsilon))
    return yd
##

def supply(price, b):
    ys = math.e**(b*price) - 1
    return ys
##

def equilibrium(a,b,epsilon):
    maxiter = 30
    toler = 10**(-6)
    plow = 0.1          # Minimum price
    phigh = 10.0        # Maximum price
    niter = maxiter

    yd = []
    ys = []
    pcur = []

    for i in range(maxiter):
        pcur.append((plow + phigh)/2 )
#    print('Current price is: %g\n' % p)
        d = demand(pcur[-1],a,epsilon)
        s = supply(pcur[-1],b)
        yd.append(d)
        ys.append(s)

        excesssupply = ys[-1] - yd[-1]      # Use .pop() to get last element in the list    

        if excesssupply > 0:
            phigh = pcur[-1]
        else:
            plow = pcur[-1]

        diff = abs(phigh - plow)

        if diff <= toler:
            niter = i
        break

    pclear = (plow + phigh)/2
    yd_final = demand(pcur[-1], a, epsilon)
    ys_final = supply(pcur[-1], b)
    excesssupply = ys_final - yd_final


    print('--------------------------------------')
    print('For (a, b, epsilon) = (%g,%g,%g)\n' % (a,b,epsilon))
    print('Number of iteration is: %i\n' % niter)
    print('Market clearing price is: %g\n' % pclear)
    print('Demand is: %g\n' % yd_final)
    print('Supply is: %g\n' % ys_final)
    print('Excess supply is: %g\n' % excesssupply)

    # Order data to plot properly
    order = np.argsort(pcur)
    pcur = np.array(pcur)[order]
    yd = np.array(yd)[order]
    ys = np.array(ys)[order]

    ##




# Main function
parms1 = (1,0.1,1)
parms2 = (2,0.1,1)
parms3 = (1,0.1,2)

equilibrium(*parms1) # using *parms1 passes the structure of args to the function
equilibrium(*parms2)
equilibrium(*parms3)

## Plot data with legends
# plt.plot(pcur,yd, 'b-', label='demand')
# plt.plot(pcur,ys, 'r-', label='supply')
#axes = plt.gca()
#axes.set_xlim([2, 4.0])
#axes.set_ylim([0.2,0.5])
#plt.legend()

#plt.show()


```
